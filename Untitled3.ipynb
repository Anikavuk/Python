```python
df.columns = ['invoice_no', 'stock_code', 'description', 'quantity', 'invoice_date',
       'unit_price', 'customer_id', 'country']
```

```python
df['invoice_date'][0]
```

```python
df['invoice_date'] = pd.to_datetime(df['invoice_date'], format='%m/%d/%Y %H:%M')
#df['invoice_date'] = pd.to_datetime(df['invoice_date'], format='%m/%d/%Y %H:%M')
```

```python
df['invoice_date'][0]
```

```python
df['invoice_month'] = pd.to_datetime(pd.DataFrame({'day': 1, 
                                              'month': df['invoice_date'].dt.month, 
                                              'year': df['invoice_date'].dt.year},
                                              index=df.index))
```

```python
df.sample(5)
```

```python
df['revenue'] = df['quantity'] * df['unit_price']
df
```

```python
t = df.groupby('invoice_month')['revenue'].agg(['count', 'sum']).reset_index()
```
